### <u> Una sÃ³la iteraciÃ³n  <u>

```python
def energia(SM,J,h,N):
	#EnergÃ­a promedio
	E = 0
	m = np.sum(SM)	 
	E = -(J/2)*(m)**2/N - h*m
	
	return E

def heat_bath(N,J,h,it,beta):
	m = []
	chi = []
	mat = []
	SM = np.random.choice([-1,1],[N])
	
	for i in range (it):
		SM_f = np.copy(SM)
		#Elige spin aleatorio que girar
		i = np.random.randint(N)
	
		SM_f[i] *= -1
		
		E_i = energia(SM,J,h,N)
		E_f = energia(SM_f,J,h,N)
	
		ÎE = E_f - E_i
		
		if ÎE<0:
			SM = np.copy(SM_f)
		if ÎE>0:
			if np.random.random()<np.exp(-ÎE*beta):
				SM = np.copy(SM_f)

		m.append(np.mean(SM))
		chi.append(beta*(np.mean(SM**2) - np.mean(SM)**2))
		mat.append(SM)
		
	return m,chi,mat
```

### 3.1. **Curie Weiss algoritmo metrÃ³polis**
- Los dos algoritmos hacen lo mismo
- Comprobamos casos para temperatura por encima y por debajo de la crÃ­tica

```python
def cambio_energia(SM,J,h,N,i):
	#EnergÃ­a promedio
	E = ((2*J)*sum(np.delete(SM,i))+2*h)*SM[i]
	return E
```

```python
def metro(N,J,h,it,beta):
	m = []
	chi = []
	mat = []
	SM = np.random.choice([-1,1],[N])
	
	for j in range (it):
		for k in range(N):
			i = np.random.randint(N)
			ÎE = cambio_energia(SM,J,h,N,i)
			if ÎE<0:
				SM[i] *= -1
			else:
				if np.random.random()<np.exp(-ÎE*beta):
					SM[i] *= -1

		m.append(np.mean(SM))
		chi.append(beta*(np.mean(SM**2) - np.mean(SM)**2))
		mat.append(SM)
		
	return m,chi,mat
```

```python
def metropolis(N,J,h,it,beta):
	m = []
	SM = np.random.choice([-1,1],[N])
	
	for j in range (it):
		for k in range(N):
			i = np.random.randint(N)
			ÎE = cambio_energia(SM,J,h,N,i)
			w = min(1,np.exp(-ÎE*beta))
			if np.random.random() < w:
				SM[i] *= -1                
		m.append(np.mean(SM))
	return m,SM
```

### 3.2. Curie - Weiss Algoritmo Heat Bath

```python
def cambio_energia(SM,J,h,N,i):
	#EnergÃ­a promedio
	E = ((2*J)*sum(np.delete(SM,i))+2*h)*SM[i]
	return E
def HB(N,J,h,it,beta):
	m = []
	chi = []
	mat = []
	SM = np.random.choice([-1,1],[N])
	
	for j in range (it):
		for k in range(N):
			i = np.random.randint(N)
			ÎE = cambio_energia(SM,J,h,N,i)
			p = 1/(1+np.exp(beta*ÎE))
			if np.random.random() < p:
					SM[i] *= -1

		m.append(np.mean(SM))
		chi.append(beta*(np.mean(SM**2) - np.mean(SM)**2))
		mat.append(SM)
		
	return m,chi,SM
```
